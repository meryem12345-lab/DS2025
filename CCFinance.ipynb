The second code snippet performs the following key actions:

**Reasoning**:
The next step is to execute Code Snippet 3, which involves creating a dictionary for population data. After execution, I will provide a summary of its purpose.



````python
print("--- Executing Code Snippet 3 ---")
# Dictionnaire des populations (en millions d'habitants)
# Source : Banque mondiale, 2024
populations_2024 = {
    'États-Unis': 337,
    'Chine': 1411,
    'Allemagne': 84,
    'Japon': 125,
    'Royaume-Uni': 67,
    'France': 65,
    'Inde': 1430,
    'Brésil': 216
}

# Ajout de la population au DataFrame existant
df['Population_2024'] = df['Pays'].map(populations_2024)

# Calcul du PIB par habitant pour 2024 (en milliers USD)
# Formule : (PIB 2024 * 1000) / Population 2024
df['PIB_par_habitant_2024'] = ((df['2024'] * 1000) / df['Population_2024']).round(2)

print("Population data added to DataFrame and GDP per capita calculated.")
print("First 5 rows of the updated DataFrame:")
print(df.head())
````

### Summary of Code Snippet 3

Code Snippet 3 extends the data analysis by:

*   **Defining Population Data**: It creates a dictionary `populations_2024` containing the estimated populations (in millions) for 2024 for each of the eight countries, sourced from the World Bank.
*   **Integrating Population into DataFrame**: It adds a new column, `Population_2024`, to the existing `df` DataFrame by mapping the population data to the corresponding countries.
*   **Calculating GDP per Capita**: It computes the 'PIB_par_habitant_2024' (GDP per capita for 2024) by dividing the GDP for 2024 (multiplied by 1000 to convert billions to millions for per capita calculation) by the respective population, rounding the result to two decimal places.

This snippet enriches the DataFrame with crucial demographic information, enabling more nuanced economic analysis.

**Reasoning**:
The next step is to execute Code Snippet 4, which involves transforming the DataFrame into a "long" format for better temporal visualization. After execution, I will provide a summary of its purpose.



````python
print("--- Executing Code Snippet 4 ---")
# Création d'une version "longue" du DataFrame pour faciliter les visualisations temporelles
# On transforme les colonnes d'années en lignes

# Sélection des colonnes d'années
annees = [str(year) for year in range(2020, 2025)]

# Utilisation de pd.melt pour "dépivoter" le DataFrame
df_long = df.melt(id_vars=['Pays', 'Population_2024', 'PIB_par_habitant_2024'],
                 value_vars=annees,
                 var_name='Annee',
                 value_name='PIB_Milliards_USD')

# Convertir la colonne 'Annee' en entier pour les graphiques
df_long['Annee'] = df_long['Annee'].astype(int)

print("DataFrame successfully transformed to long format.")
print("First 5 rows of the long DataFrame:")
print(df_long.head())
print(f"Total observations in long format: {len(df_long)}")
````

**Reasoning**:
The next step is to execute Code Snippet 5, which involves calculating the annual growth rates for each country. After execution, I will provide a summary of its purpose.



````python
print("--- Executing Code Snippet 5 ---")
# Calcul des taux de croissance annuels pour chaque pays
# Formule : ((PIB année N / PIB année N-1) - 1) * 100

# Création d'un nouveau DataFrame pour les taux de croissance
croissance = pd.DataFrame({'Pays': df['Pays']})

for i in range(2021, 2025):
    year_n = str(i)
    year_n_minus_1 = str(i - 1)
    croissance[f'Croissance_{year_n}_%'] = (((df[year_n] / df[year_n_minus_1]) - 1) * 100).round(2)

print("Annual growth rates calculated.")
print("First 5 rows of the growth rate DataFrame:")
print(croissance.head())
````

### Summary of Code Snippet 5

Code Snippet 5 calculates the annual GDP growth rates for each country over the period 2021-2024:

*   **Initialization**: It creates an empty Pandas DataFrame named `croissance` and initializes it with the `Pays` (Country) column from the main `df` DataFrame.
*   **Growth Rate Calculation**: It iterates from 2021 to 2024, calculating the percentage change in GDP from the previous year for each country using the formula `((GDP_Year_N / GDP_Year_N-1) - 1) * 100`. The results are rounded to two decimal places.
*   **DataFrame Population**: For each year, a new column (e.g., `Croissance_2021_%`) is added to the `croissance` DataFrame to store the calculated growth rates.

This snippet provides a clear view of the year-over-year economic performance of each country.

**Reasoning**:
The next step is to execute Code Snippet 8, which involves calculating market shares based on GDP. After execution, I will provide a summary of its purpose.



````python
print("--- Executing Code Snippet 8 ---")
# Calcul du PIB total mondial (pour l'échantillon)
pib_total = df['2024'].sum()

# Calcul des parts de marché
df['Part_PIB_%'] = (df['2024'] / pib_total * 100).round(1)

# Tri par part décroissante
parts_de_marche = df[['Pays', '2024', 'Part_PIB_%']].sort_values(by='Part_PIB_%', ascending=False)

print("Market shares calculated and sorted.")
print("Market share data (top 5 rows):")
print(parts_de_marche.head())
````

### Summary of Code Snippet 8

Code Snippet 8 focuses on calculating and presenting the market share of each country's GDP within the analyzed sample for the year 2024.

*   **Total GDP Calculation**: It first sums up the '2024' GDP values of all countries in the `df` DataFrame to determine the total GDP of the sample.
*   **Market Share Calculation**: It then calculates the percentage share of each country's 2024 GDP relative to this total, storing the result in a new column named `Part_PIB_%`. The percentages are rounded to one decimal place.
*   **Sorting**: Finally, it creates a new DataFrame `parts_de_marche` containing the `Pays`, `2024` GDP, and `Part_PIB_%`, and sorts it in descending order of market share.

This snippet provides insight into the relative economic size and dominance of each country within the defined group.

**Reasoning**:
The next step is to execute Code Snippet 9, which involves preparing data for correlation analysis and calculating the correlation matrix. After execution, I will provide a summary of its purpose.



````python
print("--- Executing Code Snippet 9 ---")
# Préparation des données pour l'analyse de corrélation
df_corr = df[['2020', '2021', '2022', '2023', '2024', 'Population_2024', 'PIB_par_habitant_2024']].copy()

# Calcul de la matrice de corrélation
correlation_matrix = df_corr.corr()

print("Data prepared for correlation analysis and correlation matrix calculated.")
print("Correlation Matrix:")
print(correlation_matrix)
````

### Summary of Code Snippet 9

Code Snippet 9 prepares the data and calculates the correlation matrix between various economic indicators:

*   **Data Preparation**: It creates a new DataFrame, `df_corr`, by selecting relevant columns from the main `df` DataFrame, including GDP for years 2020-2024, `Population_2024`, and `PIB_par_habitant_2024`.
*   **Correlation Matrix Calculation**: It then computes the pairwise correlation between all selected variables using the `.corr()` method. The result is stored in `correlation_matrix`.

This correlation matrix helps in understanding the linear relationships and interdependencies between the economic growth over years, population, and GDP per capita.

**Reasoning**:
The next step is to execute Code Snippet 10, which involves calculating cumulative growth and the Compound Annual Growth Rate (CAGR). After execution, I will provide a summary of its purpose.



````python
print("--- Executing Code Snippet 10 ---")
# Calcul de la croissance cumulée sur la période 2020-2024
df['Croissance_Cumulee_%'] = ((df['2024'] / df['2020']) - 1) * 100

# Calcul du Taux de Croissance Annuel Moyen (TCAM)
# Formule : ((Valeur_finale / Valeur_initiale)^(1/nombre_annees)) - 1
# Pour 5 ans (2020, 2021, 2022, 2023, 2024), il y a 4 périodes de croissance.
df['TCAM_%'] = (((df['2024'] / df['2020'])**(1/4)) - 1) * 100

print("Cumulative growth and CAGR calculated.")
print("First 5 rows of the DataFrame with new columns:")
print(df.head())
````

### Summary of Code Snippet 10

Code Snippet 10 adds two new essential metrics to the `df` DataFrame:

*   **Cumulative Growth (`Croissance_Cumulee_%`)**: This calculates the total percentage change in GDP for each country from 2020 to 2024. It indicates the overall growth (or decline) over the entire period.
*   **Compound Annual Growth Rate (CAGR) (`TCAM_%`)**: This calculates the annualized average growth rate of GDP over the period 2020-2024. It provides a smoothed, geometric mean return over the specified duration, useful for comparing growth rates across different periods or entities. The calculation uses 4 periods (2020-2021, 2021-2022, 2022-2023, 2023-2024).

**Reasoning**:
The next step is to execute Code Snippet 11, which involves creating a time-series plot of GDP for each country. After execution, I will provide a summary of its purpose.



````python
print("--- Executing Code Snippet 11 ---")
# Création de la figure avec une taille appropriée
plt.figure(figsize=(16, 9))

# Boucle pour tracer chaque pays
for pays in df['Pays']:
    # Extraction des données pour le pays
    donnees_pays = df[df['Pays'] == pays][annees].values[0]

    # Traçage de la courbe avec marqueurs
    plt.plot(annees, donnees_pays, marker='o', linewidth=2.5,
             markersize=8, label=pays, alpha=0.8)

# Configuration du graphique
plt.title('Évolution du PIB par Pays (2020-2024)', fontsize=18)
plt.xlabel('Année', fontsize=14)
plt.ylabel('PIB (Milliards USD)', fontsize=14)
plt.xticks(annees, rotation=45, fontsize=12) # Afficher toutes les années comme ticks
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Pays', fontsize=12, title_fontsize='13', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout() # Ajuste le layout pour éviter le chevauchement
plt.show()

print("Time-series plot of GDP per country generated.")
````

### Summary of Code Snippet 11

Code Snippet 11 generates a time-series plot visualizing the evolution of GDP for each country from 2020 to 2024:

*   **Figure Creation**: It initializes a matplotlib figure with a specified size for better readability.
*   **Iterative Plotting**: It loops through each country in the `df` DataFrame, extracts its GDP data for the specified years (`annees`), and plots it as a line graph with markers.
*   **Graph Configuration**: It sets the title, axis labels, custom x-axis ticks (to display all years), and adds a grid. A legend is placed outside the plot area to avoid overlap, and `tight_layout()` is used for optimal spacing.

This plot provides a clear visual representation of each country's GDP trend over the five-year period, allowing for easy comparison of economic trajectories.
